```python
from pyecharts import Bar
```

```python
attr = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
```

```python
v1 = [5, 20, 36, 10, 75, 90]
```

```python
v2 = [10, 25, 8, 60, 20, 80]
```

```python
bar = Bar("柱状图数据堆叠示例")
```

```python
bar.add("商家A", attr, v1,mark_point=["average"],mark_line=["min", "max"], is_stack=True)
```

```python
bar.add("商家B", attr, v2,mark_point=["average"],mark_line=["min", "max"], is_stack=True)
```

```python
bar.render("test.html")
```

# Bar - datazoom - slider 示例

```python
import random

attr = ["{}天".format(i) for i in range(30)]
v1 = [random.randint(1, 30) for _ in range(30)]
bar = Bar("Bar - datazoom - slider 示例")
bar.add("", attr, v1, is_label_show=True, is_datazoom_show=True)
bar.render("test1.html")
```

# Bar - datazoom - inside 示例

```python
attr = ["{}天".format(i) for i in range(30)]
v1 = [random.randint(1, 30) for _ in range(30)]
bar = Bar("Bar - datazoom - inside 示例")
bar.add("", attr, v1, is_datazoom_show=True, datazoom_type='inside', datazoom_range=[10, 25])
bar.render("test2.html")
```

# Bar - datazoom - both 示例

```python
attr = ["{}天".format(i) for i in range(30)]
v1 = [random.randint(1, 30) for _ in range(30)]
bar = Bar("Bar - datazoom - inside 示例")
bar.add("", attr, v1, is_datazoom_show=True, datazoom_type='both', datazoom_range=[10, 25])
bar.render("test3.html")
```
